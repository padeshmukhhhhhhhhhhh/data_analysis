```python
import pandas as pd
import yfinance as yf
```

```python
data=yf.Ticker("ITC.NS")
```

```python
symbol="ITC.NS"
df=data.history(symbol, start="2023-1-30", end="2024-01-30", interval='1h')
```

```python
del df["Dividends","Volume","Stock Splits"]
```

```python
df
```

```python
import talib as ta
```

```python
df["RSI"]=ta.RSI(df["Close"],timeperiod=14)
```

```python
df[14:]
```

```python
print(df["RSI"][14])
```

```python
position=False

for i in range(14,len(df)):
    if df["RSI"][i]<30 and position == False:
        df.loc[i, 'Buy'] = 1
        position=True
    elif df["RSI"][i]>70 and position == True: 
        df.loc[i, 'Sell'] = 1
        position=False
    else:
        pass
```
