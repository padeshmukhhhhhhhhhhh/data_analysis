```python
import pandas as pd
import yfinance as yf
```

```python
data=yf.Ticker("ITC.NS")
```

```python
symbol="ITC.NS"
df=data.history(symbol, start="2023-1-30", end="2024-01-30", interval='1h')
```

```python
  del df["Volume"], df["Dividends"], df["Stock Splits"]
```

```python
import talib as ta
```

```python
df["RSI"]=ta.RSI(df["Close"],timeperiod=14)
```

```python
df[14:]
```

```python
for i in range(len(df)):
    if i ==0:
        df.iloc[i]
```

```python
position=False
df["Buy"]=0
df["Sell"]=0
df["Signal"]="-"

for i in range(14,len(df)):
    if df.iloc[i,df.columns.get_loc('RSI')]<30 and position == False:
  
        df.iloc[i, df.columns.get_loc('Buy')] = 1
        df.iloc[i, df.columns.get_loc('Signal')] = "Buy"
        
        position=True
        
    elif df.iloc[i,df.columns.get_loc('RSI')]>70 and position == True: 
        
        df.iloc[i, df.columns.get_loc('Sell')] = 1
        df.iloc[i, df.columns.get_loc('Signal')] = "Sell"
        
        position=False
    else:
        pass
```
